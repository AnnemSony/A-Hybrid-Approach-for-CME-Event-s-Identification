```python
import os
import numpy as np
import pandas as pd
import cdflib

# Set data directory and variable list
data_dir = r"."
blk_vars = ['proton_density', 'proton_bulk_speed', 'proton_thermal',
            'alpha_density', 'alpha_bulk_speed', 'alpha_thermal']

# --- Load TH1/TH2 (average 50-bin flux) ---
def load_flux_avg(file_list, label):
    dfs = []
    for fname in file_list:
        cdf = cdflib.CDF(os.path.join(data_dir, fname))
        epoch = cdflib.cdfepoch.to_datetime(cdf.varget("epoch_for_cdf_mod"))
        flux = cdf.varget("integrated_flux_mod")  # shape: [N, 50]
        flux = np.where(flux > -1e30, flux, np.nan)
        avg_flux = np.nanmean(flux, axis=1)
        dfs.append(pd.DataFrame({'datetime': epoch, f'avg_flux_{label}': avg_flux}))
    return pd.concat(dfs, ignore_index=True)

# --- Load BLK Data ---
def load_blk(file_list):
    dfs = []
    for fname in file_list:
        cdf = cdflib.CDF(os.path.join(data_dir, fname))
        epoch = cdflib.cdfepoch.to_datetime(cdf.varget("epoch_for_cdf_mod"))
        df = pd.DataFrame({'datetime': epoch})
        for var in blk_vars:
            val = cdf.varget(var)
            df[var] = np.where(val > -1e30, val, np.nan)
        dfs.append(df)
    return pd.concat(dfs, ignore_index=True)

# --- Load File Lists ---
cdf_files = {
    "TH1": sorted([f for f in os.listdir(data_dir) if "TH1" in f and f.endswith(".cdf")]),
    "TH2": sorted([f for f in os.listdir(data_dir) if "TH2" in f and f.endswith(".cdf")]),
    "BLK": sorted([f for f in os.listdir(data_dir) if "BLK" in f and f.endswith(".cdf")])
}

# --- Load DataFrames ---
df_th1 = load_flux_avg(cdf_files['TH1'], "th1")
df_th2 = load_flux_avg(cdf_files['TH2'], "th2")
df_blk = load_blk(cdf_files['BLK'])

# --- Merge all DataFrames on datetime ---
df_merged = df_blk.merge(df_th1, on='datetime', how='outer')
df_merged = df_merged.merge(df_th2, on='datetime', how='outer')
df_merged.sort_values('datetime', inplace=True)

# --- Forward fill missing values up to 6 consecutive NaNs (~1 hour if 10-min cadence) ---
df_merged = df_merged.ffill(limit=6).reset_index(drop=True)

# Preview
print(df_merged.head())
```

```python
import numpy as np
import pandas as pd

# Improved compute_halo_cme_features function with robust handling
def compute_halo_cme_features(df, window=6):
    df = df.copy()
    df.set_index('datetime', inplace=True)

    # Moving averages
    df['flux_th1_ma'] = df['avg_flux_th1'].rolling(window=window, min_periods=1).mean()
    df['flux_th2_ma'] = df['avg_flux_th2'].rolling(window=window, min_periods=1).mean()
    df['speed_ma'] = df['proton_bulk_speed'].rolling(window=window, min_periods=1).mean()
    df['thermal_ma'] = df['proton_thermal'].rolling(window=window, min_periods=1).mean()
    df['density_ma'] = df['proton_density'].rolling(window=window, min_periods=1).mean()

    # Standard deviations with minimum periods
    df['speed_std'] = df['proton_bulk_speed'].rolling(window=window, min_periods=2).std()
    df['density_std'] = df['proton_density'].rolling(window=window, min_periods=2).std()

    # Flux spike score
    df['flux_spike_score'] = np.where(
        df['flux_th1_ma'] > 0,
        (df['avg_flux_th1'] - df['flux_th1_ma']) / df['flux_th1_ma'],
        0
    )

    # Speed jump score
    df['speed_grad'] = df['proton_bulk_speed'].diff().fillna(0).abs()
    df['speed_jump_score'] = np.where(
        df['speed_std'] > 0,
        df['speed_grad'] / df['speed_std'],
        0
    )

    # Thermal shock score
    df['thermal_grad'] = df['proton_thermal'].diff().fillna(0)
    df['thermal_shock_score'] = np.where(
        df['thermal_ma'] > 0,
        df['thermal_grad'] / df['thermal_ma'],
        0
    )

    # Density spike score
    df['density_spike_score'] = np.where(
        df['density_std'] > 0,
        (df['proton_density'] - df['density_ma']) / df['density_std'],
        0
    )

    # # Composite CME score
    # df['halo_cme_event_score'] = (
    #     0.3 * df['flux_spike_score'].clip(lower=0) +
    #     0.25 * df['speed_jump_score'].clip(lower=0) +
    #     0.2 * df['thermal_shock_score'].clip(lower=0) +
    #     0.2 * df['density_spike_score'].clip(lower=0)
    # )

    df.reset_index(inplace=True)
    return df

# Apply the improved feature generation
df_features_final = compute_halo_cme_features(df_cleaned)

# Display the result
# import ace_tools as tools; tools.display_dataframe_to_user(name="Improved Halo CME Feature Scores", dataframe=df_features_final)
```

```python
def add_cme_flags_from_thresholds(df, z=2.5, percentile=98):
    df = df.copy()

    # Features to evaluate for CME-related anomalies
    feature_cols = [
        'flux_spike_score',
        'speed_jump_score',
        'density_spike_score',
        'thermal_shock_score'
    ]

    # 1. Z-Score thresholds
    zscore_thresholds = {
        col: df[col].mean() + z * df[col].std()
        for col in feature_cols
    }

    # 2. Percentile thresholds
    percentile_thresholds = {
        col: df[col].quantile(percentile / 100.0)
        for col in feature_cols
    }

    # 3. Flagging based on thresholds
    for col in feature_cols:
        df[f'{col}_z_flag'] = df[col] > zscore_thresholds[col]
        df[f'{col}_p_flag'] = df[col] > percentile_thresholds[col]

    # 4. Composite Flags
    df['zscore_cme_flag'] = df[[f'{col}_z_flag' for col in feature_cols]].sum(axis=1) >= 2
    df['percentile_cme_flag'] = df[[f'{col}_p_flag' for col in feature_cols]].sum(axis=1) >= 2

    return df, zscore_thresholds, percentile_thresholds
df_flagged, z_th, p_th = add_cme_flags_from_thresholds(df_features_final)
```

```python
import numpy as np
import pandas as pd

def auto_tune_without_groundtruth(df, z_vals=[2.0, 2.5, 3.0], p_vals=[95, 98, 99], min_flags_list=[2, 3]):
    feature_cols = [
        'flux_spike_score',
        'speed_jump_score',
        'density_spike_score',
        'thermal_shock_score'
    ]
    
    results = []

    for z in z_vals:
        for p in p_vals:
            for min_flags in min_flags_list:
                # Flagging logic
                df_temp, _, _ = add_cme_flags_from_thresholds(df.copy(), z=z, percentile=p)
                z_flag_cols = [f'{col}_z_flag' for col in feature_cols]
                df_temp['tuned_cme_flag'] = df_temp[z_flag_cols].sum(axis=1) >= min_flags

                # Composite scoring
                df_temp['cme_composite_score'] = (
                    0.3 * df_temp['flux_spike_score'].clip(lower=0) +
                    0.25 * df_temp['speed_jump_score'].clip(lower=0) +
                    0.25 * df_temp['density_spike_score'].clip(lower=0) +
                    0.2 * df_temp['thermal_shock_score'].clip(lower=0)
                )

                # Extract detections
                detections = df_temp[df_temp['tuned_cme_flag']]
                num_detections = len(detections)
                avg_score = detections['cme_composite_score'].mean() if num_detections else 0

                # Scoring: few events, high quality = good
                score = (avg_score + 1) / (np.log(num_detections + 2))  # Prevent div by 0

                results.append({
                    'z': z,
                    'percentile': p,
                    'min_flags': min_flags,
                    'num_detections': num_detections,
                    'avg_composite_score': round(avg_score, 2),
                    'unsupervised_score': round(score, 4)
                })

    return pd.DataFrame(results).sort_values(by='unsupervised_score', ascending=False)
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest

# --- Composite CME Score ---
def add_composite_score(df):
    df = df.copy()
    df['cme_composite_score'] = (
        0.3 * df['flux_spike_score'].clip(lower=0) +
        0.25 * df['speed_jump_score'].clip(lower=0) +
        0.25 * df['density_spike_score'].clip(lower=0) +
        0.2 * df['thermal_shock_score'].clip(lower=0)
    )
    return df

# --- Auto-tuned Statistical Threshold Detection Only ---
def add_cme_flags_from_thresholds(df, z, p):
    df = df.copy()
    features = ['flux_spike_score', 'speed_jump_score', 'density_spike_score', 'thermal_shock_score']
    for col in features:
        z_th = df[col].mean() + z * df[col].std()
        p_th = df[col].quantile(p / 100.0)
        df[f'{col}_z_flag'] = df[col] > z_th
        df[f'{col}_p_flag'] = df[col] > p_th
    return df

# --- Auto-tuning Only: Select First Valid High Z/Percentile Detection ---
def auto_tune_stat_threshold(df, z_vals=[3.0, 2.5, 2.0], p_vals=[99, 98, 95], min_flags_list=[3, 2]):
    features = ['flux_spike_score', 'speed_jump_score', 'density_spike_score', 'thermal_shock_score']
    for z in z_vals:
        for p in p_vals:
            for min_flags in min_flags_list:
                df_tmp = df.copy()
                df_tmp = add_cme_flags_from_thresholds(df_tmp, z, p)
                z_flags = [f'{col}_z_flag' for col in features]
                df_tmp['tuned_cme_flag'] = df_tmp[z_flags].sum(axis=1) >= min_flags
                if df_tmp['tuned_cme_flag'].sum() > 0:
                    return df_tmp, (z, p, min_flags)
    raise ValueError("No CME events detected with provided z/percentile/min_flag combinations.")

# --- Isolation Forest ---
def apply_isolation_forest(df, contamination=0.005):
    X = df[['flux_spike_score', 'speed_jump_score', 'density_spike_score', 'thermal_shock_score']]
    iso = IsolationForest(contamination=contamination, random_state=42)
    df['iso_flag'] = iso.fit_predict(X) == -1
    return df

# --- Final Detection Pipeline ---
def detect_cme_combined(df):
    df = add_composite_score(df)
    df, (z, p, min_flags) = auto_tune_stat_threshold(df)
    
    features = ['flux_spike_score', 'speed_jump_score', 'density_spike_score', 'thermal_shock_score']
    df['stat_flag'] = df[[f'{col}_z_flag' for col in features]].sum(axis=1) >= min_flags

    df['phys_flag'] = (
        (df['proton_bulk_speed'] > 500) &
        (df['proton_density'] > 5) &
        (df['avg_flux_th1'] > 1e7)
    )

    df = apply_isolation_forest(df)

    df['cme_ensemble_flag'] = df[['stat_flag', 'phys_flag', 'iso_flag']].sum(axis=1) >= 2

    tuning_result = {
        'z': z,
        'percentile': p,
        'min_flags': min_flags,
        'num_ensemble_detections': int(df['cme_ensemble_flag'].sum())
    }

    return df, tuning_result

# --- Print CME Event Timestamps ---
def print_detected_cme_events(df):
    events = df[df['cme_ensemble_flag']]['datetime'].tolist()
    print(f"\n🌟 Detected CME Events (Total={len(events)}):")
    for t in events:
        print(f" - {t}")

# --- Export Detected Events ---
def export_detected_cme_events(df, filename='cme_detected_events.csv'):
    cme_df = df[df['cme_ensemble_flag']][['datetime', 'cme_composite_score']]
    cme_df.to_csv(filename, index=False)
    print(f"✅ Exported {len(cme_df)} events to {filename}")
```

```python
import matplotlib.pyplot as plt

def plot_cme_signals(df, flag_col='zscore_cme_flag', title='CME Detection (Z-Score)'):
    fig, axs = plt.subplots(4, 1, figsize=(16, 12), sharex=True)

    # Plot TH1 & TH2 flux
    axs[0].plot(df['datetime'], df['avg_flux_th1'], label='avg_flux_th1', alpha=0.7)
    axs[0].plot(df['datetime'], df['avg_flux_th2'], label='avg_flux_th2', alpha=0.7)
    axs[0].legend()
    axs[0].set_ylabel('Flux')
    axs[0].grid(True)

    # Proton speed
    axs[1].plot(df['datetime'], df['proton_bulk_speed'], label='proton_bulk_speed', color='orange')
    axs[1].legend()
    axs[1].set_ylabel('Speed (km/s)')
    axs[1].grid(True)

    # Proton density
    axs[2].plot(df['datetime'], df['proton_density'], label='proton_density', color='green')
    axs[2].legend()
    axs[2].set_ylabel('Density (cm⁻³)')
    axs[2].grid(True)

    # Thermal
    axs[3].plot(df['datetime'], df['proton_thermal'], label='proton_thermal', color='red')
    axs[3].legend()
    axs[3].set_ylabel('Thermal')
    axs[3].grid(True)

    # Mark CME detection times
    for ax in axs:
        for time in df[df[flag_col]]['datetime']:
            ax.axvline(x=time, color='black', linestyle='--', alpha=0.3)

    plt.suptitle(title)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```

```python
import pandas as pd
import numpy as np

def detect_flux_outliers(df, window=24, z_thresh=5):
    """
    Identifies outliers in avg_flux_th1 and avg_flux_th2 using rolling Z-score.
    
    Parameters:
    - df: DataFrame containing datetime, avg_flux_th1, avg_flux_th2
    - window: rolling window size for mean/std
    - z_thresh: Z-score threshold to flag outliers
    
    Returns:
    - df_outliers: DataFrame with z-scores and flagged outliers
    """
    df = df.copy()
    df['datetime'] = pd.to_datetime(df['datetime'])

    for col in ['avg_flux_th1', 'avg_flux_th2']:
        rolling_mean = df[col].rolling(window=window, min_periods=1).mean()
        rolling_std = df[col].rolling(window=window, min_periods=1).std()
        z_col = f'z_{col}'
        outlier_col = f'{col}_is_outlier'

        df[z_col] = (df[col] - rolling_mean) / rolling_std
        df[outlier_col] = df[z_col].abs() > z_thresh

    return df

#Example usage:
df_with_outliers = detect_flux_outliers(df_features_final, window=24, z_thresh=5)

#To filter out only TH1 or TH2 outliers:
th1_outliers = df_with_outliers[df_with_outliers['avg_flux_th1_is_outlier']]
th2_outliers = df_with_outliers[df_with_outliers['avg_flux_th2_is_outlier']]
```

```python
def apply_thresholds(df):
    df['density_z'] = (df['proton_density'] - df['proton_density'].mean()) / df['proton_density'].std()
    df['speed_z'] = (df['proton_bulk_speed'] - df['proton_bulk_speed'].mean()) / df['proton_bulk_speed'].std()
    df['flux_z'] = (df['avg_flux_th1'] - df['avg_flux_th1'].mean()) / df['avg_flux_th1'].std()
    
    # Flag outliers
    df['density_flag'] = df['density_z'].abs() > 2.5
    df['speed_flag'] = df['speed_z'].abs() > 2.5
    df['flux_flag'] = df['flux_z'].abs() > 3.0

    # Composite event detection (at least 2 features spike)
    df['cme_detected'] = (df[['density_flag', 'speed_flag', 'flux_flag']].sum(axis=1) >= 2)
    return df
```
